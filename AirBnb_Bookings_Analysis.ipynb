Write your code to make your dataset analysis ready.

````python
airbnb_df.dropna(subset = ['host_name'], inplace=True)

#Removing rows with nan value for host_name as it cannot be replaced with suitable values
````

````python
airbnb_df.dropna(subset = ['name'], inplace=True)

#Removing rows with nan value for name as it cannot be replaced with suitable values
````

````python
airbnb_df['last_review'].fillna(airbnb_df['last_review'].mode()[0],inplace=True)

#Replaced with mode value as the most reviwed day is the most apporprite value for the missing ones
````

````python
airbnb_df['reviews_per_month'].fillna(airbnb_df['reviews_per_month'].mean(),inplace=True)

#Replaced with mean value of the column as it is appropriate to provide averages of the column in the missing ones
````

### What all manipulations have you done and insights you found?

The following are the manipulations conducted to clean the data and make it ready for proper analysis.  

1.   Removed rows with NaN value for 'host_name' as it cannot be replaced with suitable values
2.   Removed rows with nan value for 'name' as it cannot be replaced with suitable values
3.   Replaced 'last_review' with mode value as the most reviwed day is the most apporprite value for the missing ones
4.   Replaced 'reviews_per_month' with mean value of the column as it is appropriate to provide averages of the column in the missing ones

It can be observed after the manipulations the number of rows have been lowered by (16+21) = 37 as these rows has to be deducted. The data has been previously of 48895 rows and it is lowered to 48858 in the later sections.

It is also observed that the missing values for 'last_review' and 'reviews_per_month' has been filled with mode and mean values of the respective columns.

The dataset is now ready for analysis and the deducted rows being much less than the original rows, will not affect the outcomes of the data.
